Import data

```python
import csv

with open('data.csv', 'r') as f:
    reader = csv.reader(f)

    documents = []
    metadatas = []
    ids = []
    id = 1

    for i, row in enumerate(reader):
        if i == 0:
            continue

        documents.append(row[1])
        metadatas.append({"item_id": row[0]})
        ids.append(str(id))
        id += 1
        
```

```python
documents
```

Get the chroma client

```python
import chromadb
from chromadb.utils import embedding_functions

# chroma_client = chromadb.Client()
chroma_client = chromadb.PersistentClient(path="bookdb")
```

Create a collection

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-mpnet-base-v2")
collection = chroma_client.create_collection(name="book_collection", embedding_function=sentence_transformer_ef)
```

Add some text documents to the collection

```python
collection.add(
        documents=documents,
        metadatas=metadatas,
        ids=ids
    )
```

Query the collection

```python
results = collection.query(
    query_texts=["art"], # Chroma will embed this for you
    n_results=5, # how many results to return
    # include=['distances', 'metadatas', 'documents']
    include=['documents']
)

results
```
